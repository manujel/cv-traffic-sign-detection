```python
print("Exists:", GTSDB_ROOT.exists())
print("Is dir:", GTSDB_ROOT.is_dir())
print([p.name for p in GTSDB_ROOT.iterdir()][:10])
```

```python
candidates = sorted([p for p in GTSDB_ROOT.iterdir() if p.suffix.lower() in {".txt", ".csv"}])
print("Candidates in root:", [p.name for p in candidates])

hits = []
for pat in ["*gt*", "*annot*", "*label*", "*.txt", "*.csv"]:
    hits.extend(GTSDB_ROOT.rglob(pat))

hits = sorted(set(hits))

print("Some matches:", [str(p.relative_to(GTSDB_ROOT)) for p in hits[:20]])
print("Total matches:", len(hits))
```
